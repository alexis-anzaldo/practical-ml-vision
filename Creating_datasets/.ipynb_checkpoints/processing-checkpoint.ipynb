### Reading images

```python
import tensorflow as tf
import matplotlib.pyplot as plt

# Set the path to your image folder
data_dir = r"C:\Users\User\Desktop\GithubProjects\practical-ml-vision-book\Creating_datasets\Pistachio_Image_Dataset\Pistachio_Image_Dataset"

# Create the dataset without resizing
dataset = tf.data.Dataset.list_files(data_dir + '/*/*')

# Define a function to load and preprocess the images without resizing
def load_and_preprocess_image(file_path):
    """
    Load and preprocess an image without resizing.
    
    Parameters:
        file_path (tf.Tensor): A tensor containing the file path of the image.
        
    Returns:
        tf.Tensor: The preprocessed image tensor.
    """
    # Ensure the file path is in string format
    file_path = tf.strings.as_string(file_path)
    
    # Read the file content for each file path individually
    img = tf.io.read_file(file_path)
    
    # Decode the image to a tensor with 3 channels (RGB)
    img = tf.image.decode_image(img, channels=3)

    # Convert the pixel values to the range [0, 1] using tf.float32
    img = tf.image.convert_image_dtype(img, tf.float32)
    
    return img

# Apply the preprocessing function to each file path in the dataset
dataset = dataset.map(load_and_preprocess_image)

# Display the first 5 images from the dataset
f, ax = plt.subplots(1, 5, figsize=(15, 15))
for idx, img in enumerate(dataset.take(5)):
    # Convert the normalized image back to the 0-255 range for display
    ax[idx].imshow((img.numpy() * 255).astype(int))
    
    # Set the title of each subplot to the dimensions of the image
    ax[idx].set_title('{}x{}'.format(img.shape[0], img.shape[1]))

plt.show()
```

### Example extracting labels

```python
import tensorflow as tf
import matplotlib.pyplot as plt
import os

# Set the path to your image folder
data_dir = r"C:\Users\User\Desktop\GithubProjects\practical-ml-vision-book\Creating_datasets\Pistachio_Image_Dataset\Pistachio_Image_Dataset"

# Create a list of class names based on subdirectories
class_names = sorted([dir_name for dir_name in os.listdir(data_dir) if os.path.isdir(os.path.join(data_dir, dir_name))])

# Create the dataset without resizing
train_dataset = tf.data.Dataset.list_files(data_dir + '/*/*')

# Define a function to extract labels from file paths
def extract_label(file_path):
    parts = tf.strings.split(file_path, os.path.sep)
    # The label is the index of the class name in the list of class names
    return tf.argmax(tf.cast(tf.equal(class_names, parts[-2]), tf.int32))

# Define a function to load and preprocess the images without resizing
def load_and_preprocess_image(file_path):
    file_path = tf.strings.as_string(file_path)
    img = tf.io.read_file(file_path)
    img = tf.image.decode_image(img, channels=3)
    img = tf.image.convert_image_dtype(img, tf.float32)
    return img

# Apply the preprocessing function and extract labels to each file path in the dataset
train_dataset = train_dataset.map(lambda x: (load_and_preprocess_image(x), extract_label(x)))

# Display the first 5 images from the dataset
f, ax = plt.subplots(1, 5, figsize=(15, 15))
for idx, (img, label_int) in enumerate(train_dataset.take(5)):
    ax[idx].imshow((img.numpy() * 255).astype(int))
    ax[idx].set_title('{}x{}, \nLabel: {}'.format(img.shape[0], img.shape[1], class_names[label_int]))

plt.show()
```
